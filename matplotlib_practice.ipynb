```python
# Modifying the graph with colors, widths and markers

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston', color='green', linewidth = 2, marker = "D")
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 55)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()

# Add a grid
plt.grid()
```

```python
# Create a vertical bar chart using OOI

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

# Create the plt with ax.plot()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color='k',label='New York')
```

```python
# Create a horizontal bar chart using OOI

# Create the plot with ax.plt()
fig, ax = plt.subplots()
# Change the color to cyan
ax.barh(x_axis, y_axis, label='Chicago', color = 'cyan')
# Invert the y Axis to arrange the months from top to bottom
ax.invert_yaxis()
ax.set_title('Pyber Far by Month')
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.legend()
```

```python
# Create a Scatter Plot Using the MATLAB Method

# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

plt.scatter(x_axis, y_axis, label='Chicago', c ='red')
# Label Axes
plt.xlabel("Fare($)")
plt.ylabel("Date")

# Invert the months on y-axis
plt.gca().invert_yaxis()

# Add Title
plt.title("Pyber Fare by Month")
plt.legend()
```

```python
# Create a Bubble Chart Using the MATLAB Method

# plt.scatter(x_axis, y_axis, s=y_axis)

# multiply each data point in the y-axis by 3 using for loop/list

y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)

# Using list comprehension to replace the for loop

plt.scatter(x_axis, y_axis, s= [i*3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface

fig, ax =plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis, c="skyblue")

# Set Labels
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

# Set Title
ax.set_title('Pyber Far by Month')

# Change the size of the markers to 5 times each data point.
# change markers to 'skyblue', edgecolor to black, color transparency 20%
ax.scatter(x_axis, y_axis, 
           s= [i*5 for i in y_axis], c='skyblue', 
           alpha= 0.8, edgecolor = 'black', 
           linewidth = 2, label = 'Boston')

# Add a legend
ax.legend()

# Add a limit for the x-axis data.
ax.set_xlim(-1,11.5)

```

```python
# Create Pie Charts (5.1.8)

#Create a Pie Chart Using the MATLAB Method

plt.pie(y_axis, labels=x_axis)
```

```python
#create the pie chart with months as labels
plt.subplots(figsize=(8, 8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# Create a Pie Chart Using the Object-Oriented Interface

#Set colors
colors = ["lightcyan", "azure", "palegreen", "darkorange", "purple", 
          "lavender", "lightyellow", "plum", "gold", "indigo", "brown", 
          "pink"]
# explode 2 largest %
explode = [0,0,0.2,0,0,0,0.2,0,0,0,0,0]


# Add %, colors, shadow and explode
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis, colors = colors, 
       autopct = '%.1f%%', shadow=True, 
       explode = explode, startangle =100, counterclock = False)

#use startangle and counterclock to bring Jan to the top

# Increase the figure size to 8x8
fig.set_size_inches(8,8)


plt.show()
```
